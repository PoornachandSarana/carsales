# Finding which type of fuel cars are selling most

```python
car_df.groupby('fuel')['sold'].value_counts()
```

```python
car_df.query('sold=="Y"').groupby('fuel')['sold'].value_counts()
```

```python
car_df['sale']=(car_df['sold']=='Y').map({True:1,False:0})
```

```python
car_df.head()
```

```python
f=car_df.query('sale==1').groupby('fuel').agg({'sale':np.sum,'selling_price':np.mean,'km_driven':np.mean,})
```

```python
print(f)
```

```python
f['percent']=100*f['sale']/f['sale'].sum()
```

```python
f
```

```python
f.sort_values('percent',ascending=False)
```

It is clear that more percent of Diesel cars were selling compared to other fuel types
